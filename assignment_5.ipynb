```python
import pandas as pd
import numpy as np
import io
import requests
```

pull in the aiport location data and weather data from Github and use it to create the dataframe for the aiports and weather.

```python
url = 'https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/airports.csv'
url2 = 'https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/weather.csv'

airport_data = requests.get(url).content
airports = pd.read_csv(io.StringIO(airport_data.decode('utf-8')))

weather_data = requests.get(url2).content
weather = pd.read_csv(io.StringIO(weather_data.decode('utf-8')))
```

Question 3:
On February 12th, 2013, which New York area airport had the windiest weather?
 
Assuming:
There are no units on the speeds so it will be assumed these are in miles per hour. 
Also, I am assuming we are looking at wind speed and not wind gust.

Narative:
We will select only the rows that are for the date we are interested in.
Then we sort the wind speed column from highest value to lowest.
The first entry from ERW does not seem realistic at 1048.36.
According to a google search the highest wind ever recorded was in Tropical Cyclone Olivia with wind gust of 253 mph.
So the windest airport for this day would be LGA.

Answer 3:
As shown the airport with the hihgest wind speed on February 12, 2013 was LGA with a speed of 23.01560.

```python
weather_date = weather.loc[(weather['month'] == 2) & (weather['day'] == 12) & (weather['year'] == 2013)]                                                     
sort_wind_speed = weather_date.sort_values('wind_speed', ascending=False)
sort_wind_speed.head()
```
